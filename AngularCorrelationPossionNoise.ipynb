```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.linspace(np.pi, np.pi*9, 1000)
intensity = (np.sin(points)+1)*10
random_poission = np.random.poisson(intensity)
```

```python
plt.plot(intensity)
plt.show()
plt.plot(random_poission)
plt.show()
plt.plot(intensity+random_poission)
```

```python
def correlation(intensity):
    int_fft = np.fft.rfft(intensity)
    mean_int = np.mean(intensity)**2
    print(mean_int)
    cor = (np.fft.irfft(int_fft*np.conjugate(int_fft))/len(intensity)-np.mean(intensity)**2)/np.mean(intensity)**2
    return cor
```

```python
plt.plot(correlation(intensity))
```

```python
plt.plot(correlation(intensity+random_poission))
```
