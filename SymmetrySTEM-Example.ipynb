This notebook is designed to walk through some symmetry extraction from a dataset.  This is likely to change with time as these methods are quite new and there is a lot to explore here.  It is possible that evetually this functionality will move from pyXEM but for now it is just included with the general package


```python
import pyxem
import hyperspy.api as hs
import glob
```

```python
hs.print_known_signal_types()
```

```python
positions = sorted(glob.glob("20200930_PdNWs/ser/wave*"))
print(positions)
```

```python
pos0 = hs.load(positions[0], signal_type="DiffractionSignal")
pos0.mask_below(200)
```

```python
pos0.plot()
pos0.determine_ellipse(plot=True)
```

```python
ps =pos0.calculate_polar_spectrum()
```

```python
cor = ps.autocorrelation()
```

```python
ps.plot()
```

```python
cor.set_signal_type("correlation")
```

```python
sym = cor.get_symmetry_stem(angular_range=.01)
```

```python
sym.isig[:,2.0:].plot()
```

```python
sym.plot_symmetries(tight_layout=True, colorbar="single")
```
